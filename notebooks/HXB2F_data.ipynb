```python
def adjust_start_position(row):
  if row['Reference_sequence_name'] == 'HXB2F': return row['Start_position']
  if (row['Reference_sequence_name'] == 'chr9') and (row['dist_from_hiv']<0): return row['dist_from_hiv']
  if (row['Reference_sequence_name'] == 'chr9') and (row['dist_from_hiv']>=0): return row['dist_from_hiv'] + 9718

table_5['adjust_start_position']=table_5.apply(adjust_start_position,axis=1)
```

```python
sns.scatterplot(data=table_5, x='adjust_start_position', y='Read_coverage_at_reference', hue='Modification_type', alpha=0.5)

plt.xlabel('HXB2F')
plt.xlim(-1000, 9717+1000)
plt.title('Read Coverage in HXB2F and Adjacent Positions')
plt.show()
```

```python
from Bio import SeqIO

max_sequence_id = None
max_gc_content = 0
for seq_record in SeqIO.parse("hxb2f-grna-labeled.gb", "gb"):
    sequence = str(seq_record.seq) # convert sequence to string
    sequence_id = seq_record.id
    gc_content = (sequence.count("C") + sequence.count("G")) / len(sequence) * 100
    if gc_content > max_gc_content:
        max_sequence_id = sequence_id
        max_gc_content = gc_content

print(max_sequence_id)
print(max_gc_content)
```

```python
def get_reference():
  return next(SeqIO.parse("hxb2f-grna-labeled.gb", "gb")) 
```

```python
ref = get_reference()
ref
```

```python
import pandas as pd
```

```python
ref_series = pd.Series(list(str(ref.seq)))

next_ref_series = pd.Series(list(str(ref.seq))[1:]+['*'])
```

```python
ref_series_value = ref_series.value_counts()
ref_series_value
```

```python
is_GC = (ref_series == 'C') | (ref_series == 'G')
is_GC 
```
